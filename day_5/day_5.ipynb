```python
def parse_input(input_):
    ranges, ingredients = input_.split("\n\n")
    ingredients = set(map(int, ingredients.split()))
    ranges = [tuple(map(int, r.split("-"))) for r in ranges.split()]
    return ranges, ingredients
```

```python
def solve(input_):
    ranges, ingredients = parse_input(input_)
    max_id = max(max(ingredients), max([elem[1] for elem in ranges]))
    freshness = np.zeros(max_id + 1)
    for (b, e) in ranges:
        freshness[b:e + 1] = 1
    return sum(freshness[list(ingredients)]).astype(int)
```
