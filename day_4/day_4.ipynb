```python
test_input = """
..@@.@@@@.
@@@.@.@.@@
@@@@@.@.@@
@.@@@@..@.
@@.@@@@.@@
.@@@@@@@.@
.@.@.@.@@@
@.@@@.@@@@
.@@@@@@@@.
@.@.@@@.@."""
```

```python
import numpy as np
```

```python
from io import StringIO
```

```python
a = np.genfromtxt(StringIO(test_input.replace(".", "0 ").replace("@", "1 ")), dtype=int)  
a
```

```python
a.shape
```

```python
from scipy.signal import convolve2d
```

```python
k = np.ones(shape=(3, 3))
#k[1,1] = 0
k
```

```python
result = np.multiply(convolve2d(a, k)[1:-1,1:-1], a).astype(int)
result
```

```python
import pandas as pd
```

```python
print(pd.DataFrame(result)\
    .map(lambda x: "." if x == 0 else "x" if x <= 4 else "@")\
    .to_csv(sep=" ", index=False, header=False)\
    .replace(" ", ""))
```

```python
def mark_accessible_rolls(rolls_map):
    a = np.genfromtxt(StringIO(rolls_map.replace(".", "0 ").replace("@", "1 ")), dtype=int)
    result = np.multiply(convolve2d(a, k)[1:-1,1:-1], a).astype(int)
    #print(f"{a.shape=} {result.shape=}")
    #print(result)
    accessible_rolls_map = pd.DataFrame(result)\
        .map(lambda x: "." if x == 0 else "x" if x <= 4 else "@")\
        .to_csv(sep=" ", index=False, header=False)\
        .replace(" ", "").replace("\r", "")
    return accessible_rolls_map.strip()
```

```python
m = mark_accessible_rolls(test_input)
print(m)
m.count("x")
```
