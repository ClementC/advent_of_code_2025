```python
def solve(input_):
    start_position = input_.split("\n")[0].index("S")
    beams = {start_position}
    splits = 0
    for i, row in enumerate(input_.split("\n")[2:]):
        if i % 2 == 0:
            new_beams = set()
            for j, char in enumerate(row):
                if char == "^" and j in beams:
                    new_beams = new_beams.union({j-1, j+1})
                    splits += 1
                if char == "." and j in beams:
                    new_beams.add(j)
            beams = new_beams.copy()
    print(f"{beams=} {splits=}")
    return beams, splits
```

```python
from collections import defaultdict
```

```python
def solve_2(input_):
    start_position = input_.split("\n")[0].index("S")
    beams = {start_position}
    splits = 0
    timelines = defaultdict(int)
    timelines[start_position] = 1
    for i, row in enumerate(input_.split("\n")[2:]):
        if i % 2 == 0:
            new_row = list(row)
            new_beams = set()
            new_timelines = defaultdict(int)
            for j, char in enumerate(row):
                if char == "^" and j in beams:
                    new_beams = new_beams.union({j-1, j+1})
                    splits += 1
                    new_timelines[j-1] += timelines[j]
                    new_timelines[j+1] += timelines[j]
                if char == "." and j in beams:
                    new_beams.add(j)
                    new_timelines[j] += timelines[j]
            beams = new_beams.copy()
            for b in beams:
                new_row[b] = "|"
            timelines = new_timelines.copy()
            #print(f"{row=} {"".join(new_row)=} {timelines=} {sum(timelines.values())=}")
    print(f"{beams=} {splits=} {timelines=} {sum(timelines.values())=}")
    #return beams, splits
```
