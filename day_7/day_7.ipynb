```python
def solve(input_):
    start_position = input_.split("\n")[0].index("S")
    beams = {start_position}
    splits = 0
    for i, row in enumerate(input_.split("\n")[2:]):
        if i % 2 == 0:
            new_beams = set()
            for j, char in enumerate(row):
                if char == "^" and j in beams:
                    new_beams = new_beams.union({j-1, j+1})
                    splits += 1
                if char == "." and j in beams:
                    new_beams.add(j)
            beams = new_beams.copy()
    print(f"{beams=} {splits=}")
    return beams, splits
```

```python
from collections import defaultdict
```

```python
def solve_2(input_):
    start_position = input_.split("\n")[0].index("S")
    beams = {start_position}
    splits = 0
    timelines = defaultdict(int)
    timelines[start_position] = 1
    for i, row in enumerate(input_.split("\n")[2:]):
        if i % 2 == 0:
            new_row = list(row)
            new_beams = set()
            new_timelines = defaultdict(int)
            for j, char in enumerate(row):
                if char == "^" and j in beams:
                    new_beams = new_beams.union({j-1, j+1})
                    splits += 1
                    new_timelines[j-1] += timelines[j]
                    new_timelines[j+1] += timelines[j]
                if char == "." and j in beams:
                    new_beams.add(j)
                    new_timelines[j] += timelines[j]
            beams = new_beams.copy()
            for b in beams:
                new_row[b] = "|"
            timelines = new_timelines.copy()
            #print(f"{row=} {"".join(new_row)=} {timelines=} {sum(timelines.values())=}")
    print(f"{beams=} {splits=} {timelines=} {sum(timelines.values())=}")
    return beams, splits, timelines
```

```python
def solve_3(input_):
    row = input_.split("\n")[0]
    start_position = row.index("S")
    beams = {start_position}
    splits = 0
    timelines = defaultdict(int)
    timelines[start_position] = 1
    timelines_at_level = {0: ({7: 1}, (row.replace("S", "|"), row))}
    for i, row in enumerate(input_.split("\n")[2:]):
        if i % 2 == 0:
            new_row = list(row)
            new_beams = set()
            new_timelines = defaultdict(int)
            for j, char in enumerate(row):
                if char == "^" and j in beams:
                    new_beams = new_beams.union({j-1, j+1})
                    splits += 1
                    new_timelines[j-1] += timelines[j]
                    new_timelines[j+1] += timelines[j]
                if char == "." and j in beams:
                    new_beams.add(j)
                    new_timelines[j] += timelines[j]
            beams = new_beams.copy()
            for b in beams:
                new_row[b] = "|"
            timelines = new_timelines.copy()
            timelines_at_level[i + 2] = (timelines.copy(), ("".join(new_row), row))
            #print(f"{row=} {"".join(new_row)=} {timelines=} {sum(timelines.values())=}")
    #print(f"{beams=} {splits=} {timelines=} {sum(timelines.values())=}")
    return timelines_at_level
```

```python
def print_colored_lines(lines, line_weights, compression=False):
    """
    lines: list[str]
    line_colors: list[list[str]]  # same length as lines, each is list of colors
    
    Each color can be any Rich style: "red", "bold blue", "#ff8800", "rgb(100,200,50)", etc.
    """
    if len(lines) != len(line_weights):
        raise ValueError("lines and line_colors must have the same length")

    previous_beams = lines[0]
    previous_weigths = line_weights[0].copy()
    items_to_print = [] # [Text(lines[0].replace(".", " ").replace("|", "S")), Text(lines[0].replace(".", " "))]
    for text, weigths in zip(lines, line_weights):
        rich_text = Text()
        rich_text_2 = Text()
        processed_weigths = weigths.copy()
        if compression:
            for k,v in processed_weigths.items():
                processed_weigths[k] = 1 + math.log(v + 1)  #math.sqrt(v)
        total_w = sum(processed_weigths.values())
        
        for i, ch in enumerate(text):
            max_value = 240
            if compression:
                value = min(max_value, int(255 * (1 - processed_weigths.get(i, 0) / total_w)))
            else:
                value = min(max_value, int(255 * (1 - processed_weigths.get(i, 0) / total_w)))
            color = f"rgb({value},{value},{value})"
            rich_text.append(ch.replace(".", " ").replace("^", " "), style=color)

        for i, ch in enumerate(text):
            if compression:
                value = min(max_value, int(255 * (1 - previous_weigths.get(i, 0) / total_w)))
            else:
                value = min(max_value, int(255 * (1 - previous_weigths.get(i, 0) / total_w)))
            color = f"rgb({value},{value},{value})"
            rich_text_2.append(ch.replace(".", " ").replace("|", " "), style=color)
        
        previous_beams = text
        previous_weigths = processed_weigths.copy()
        items_to_print.append(rich_text_2)
        items_to_print.append(rich_text)

    console.print(Group(*items_to_print))
```
